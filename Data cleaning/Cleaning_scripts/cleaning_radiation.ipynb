```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\Code\miniproject_2\Direct_solar_radiation.csv')
df.head(5)
```

```python
df.drop('Observation station', axis = 1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Direct solar radiation mean [W/m2]'].str.contains('-').sum()
```

change the date to date form

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
```

```python
df['Direct solar radiation mean [W/m2]'] = pd.to_numeric(df['Direct solar radiation mean [W/m2]'], errors='coerce')
```

modify the missing data

```python
df_2012 = df[df['year'] == 2012].copy()
df_2012.head(5)
```

```python
df_2013_nan = df[(df['year'] == 2013) & (df['Direct solar radiation mean [W/m2]'].isna())].copy()
df_2013_nan.head(5)
```

```python
for i, row in df_2013_nan.iterrows():
    corresponding_2012_value = df_2012[df_2012['Date'] == row['Date']]['Direct solar radiation mean [W/m2]']
    
    if not corresponding_2012_value.empty:
        df.loc[(df['Date'] == row['Date']) & (df['Year'] == 2012), 'Direct solar radiation mean [W/m2]'] = corresponding_2012_value.iloc[0]
    
        
```

```python
df[df['year'] == 2013].sample(30)
```

calculate the mean

```python
monthly_means = df.groupby(['year', 'month'])['Direct solar radiation mean [W/m2]'].mean().reset_index(name='monthly_mean')
```

```python
df = pd.merge(df, monthly_means, on=['year', 'month'], how='left')
```

```python
df = df.drop(columns=['year', 'month'])
```
